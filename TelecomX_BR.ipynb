#🔧 Transformação

```python
#Normalizando a tabela
import pandas as pd
import requests
import json

url = requests.get('https://raw.githubusercontent.com/brendampa/Projeto_Analise_Dados_Python_TelecomXBR/refs/heads/main/TelecomX_Data.json')

df = json.loads(url.text)

pd.DataFrame (df)
```

```python
df_normalizado = pd.json_normalize(df,sep='.')
df = df_normalizado
```

```python
df
```

```python
#Verificando se o type de cada coluna está correto
df.info()
```

```python
#Transformando o dtype da coluna Charges.Total para float64 com numpy

import numpy as np

df['account.Charges.Total'] = df['account.Charges.Total'].astype(np.float64)
```

```python
#Ao tentar converter com a Numpy, o erro "ValueError: could not convert string to float: ' ' " apareceu.
#Outra forma de tentar converter com a Pandas:

df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'].str.replace(',','.'), errors='coerce') #O coerce insere o NaN quando encontra valores que não podem ser convertidos
```

```python
df.info()
```

```python
#Transformação de dados para lower
df_colunas_texto = ['customerID', 'Churn', 'customer.gender',
       'customer.Partner', 'customer.Dependents',
       'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService',
       'internet.OnlineSecurity', 'internet.OnlineBackup',
       'internet.DeviceProtection', 'internet.TechSupport',
       'internet.StreamingTV', 'internet.StreamingMovies', 'account.Contract',
       'account.PaperlessBilling', 'account.PaymentMethod']
```

```python
df[df_colunas_texto] = df[df_colunas_texto].apply(lambda x: x.str.lower())
```

```python
df.head()
```

```python
#Verificando se existem valores nulos nas colunas

df.isnull().sum()
```

```python
#Verificando se existe alguma coluna com valores vazios

for coluna in df.columns:
  if df[coluna].dtype == 'object':
    print(f'\nColuna: {coluna}')
    print(df[coluna].value_counts())
```

```python
#A coluna Churn tem 224 valores vazios. Alterando para NAN antes de transformar para "no"
df['Churn'] = df['Churn'].replace(r'^\s*$', np.nan, regex=True)
```

```python
#Transformando para "no"
df['Churn'] = df['Churn'].fillna('no')
```

```python
#Verificando se existe alguma coluna com valores vazios novamente
df['Churn'].value_counts()
```

```python
#Criação da coluna Contas_diarias, utilizando a coluna de Faturamento mensal

df['Contas_diarias'] = df['account.Charges.Monthly']/30
```

```python
df.head()
```

```python
#Transformação dos dados "yes" para 1 e "no" para 0
df_col_yes_no = ['Churn',
       'customer.Partner', 'customer.Dependents',
       'phone.PhoneService', 'phone.MultipleLines',
       'internet.OnlineSecurity', 'internet.OnlineBackup',
       'internet.DeviceProtection', 'internet.TechSupport',
       'internet.StreamingTV', 'internet.StreamingMovies',
       'account.PaperlessBilling']
```

```python
df[df_col_yes_no] = df[df_col_yes_no].replace({'yes':1,'no':0})
```

```python
df.head()
```

```python
#Renomear colunas

df.rename(columns={
    'customerID':'ClienteID', 'customer.gender':'Genero', 'customer.SeniorCitizen': 'Idoso',
       'customer.Partner':'Parceiro', 'customer.Dependents':'Dependentes', 'customer.tenure':'Tempo_contrato',
       'phone.PhoneService':'Servico_Telefone', 'phone.MultipleLines':'Multiplas_linhas', 'internet.InternetService':'Servico_internet',
       'internet.OnlineSecurity':'Seguranca_online', 'internet.OnlineBackup':'Backup_online',
       'internet.DeviceProtection':'Protecao_device', 'internet.TechSupport':'Suporte_internet',
       'internet.StreamingTV':'Streaming_tv', 'internet.StreamingMovies':'Streaming_filme', 'account.Contract':'Contrato',
       'account.PaperlessBilling':'Fatura_online', 'account.PaymentMethod':'Metodo_pagamento',
       'account.Charges.Monthly':'Fatura_mensal', 'account.Charges.Total':'Fatura_anual'
}, inplace=True)
```

```python
#Gráfico de pizza para verificar a evasão geral

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Criando variáveis
percent = df['Churn'].value_counts()
labels = ['N_evasão','Evasão']

#Aplicando um tema
sns.set_theme(style='dark')

#Criação do gráfico de pizza
fig, ax = plt.subplots(figsize=(5,5))
ax.pie(percent,
       labels=labels,
       autopct = '%.2f%%', #autopct define casas decimais
       startangle =90, #startangle controla o ângulo inicial  que começa a primeira fatia do gráfico
       textprops={'fontsize': 14},
       colors = ('#E1523D','#C2BB00')
       )
ax.set_title('Evasão de clientes - Churn', fontsize = 18)
ax.axis('equal') #Mantém o formato circular

#Salvando a figura para uso no relatório final
plt.savefig('percentual_churn.png',transparent=False, dpi=300, bbox_inches = 'tight')

plt.show()
```

```python
#Gráfico de Churn por gênero

import matplotlib.pyplot as plt
import seaborn as sns

#Criando variáveis
genero_df = df.groupby(['Genero', 'Churn']).size().reset_index(name='Quantidade') #o size calcula a quantidade

#Aplicando um tema
sns.set_theme(style='dark')

#Criação do gráfico de barras
fig, ax = plt.subplots(figsize=(8,4))
sns.barplot( data=genero_df, x='Genero', y='Quantidade', hue='Churn', #Hue segmenta a barra de acordo com a variável Churn.
                  palette='viridis',
                  errorbar=None,
                  ax=ax

       )
ax.set_title('Churn por gênero', fontsize = 18)
ax.set_xlabel('Gênero', fontsize=14)
ax.set_ylabel('Quantidade de pessoas')

#Salvando a figura para uso no relatório final
plt.savefig('Churn_por_gênero.png',transparent=False, dpi=300, bbox_inches = 'tight')

plt.show()
```

```python
#Evasão de clientes por tempo de contrtao em %

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Definir o tema escuro
sns.set_theme(style='whitegrid')

# Criar a coluna agrupadando os meses da coluna "tempo_contrato"
df['Faixa_tempo_contrato'] = pd.cut(
    df['Tempo_contrato'],
    bins=range(0, df['Tempo_contrato'].max() + 3, 3),
    right=False,
    include_lowest=True
)

# Criar figura
fig, ax = plt.subplots(figsize=(18, 10))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
ax.set_title('Evasão de clientes por Tempo de contrato', fontsize=20)
ax.set_xlabel('')
ax.set_ylabel('% de Churn',fontsize=14)
ax.set_ylim(0, 70)

sns.barplot(data=df, x='Faixa_tempo_contrato', y='Churn', ax=ax, estimator=lambda x: (x==1).mean()*100,
                errorbar=None, #Remove linhas verticais das barras (confiabilidade)
                palette='viridis',
                width=0.7, #largura barras
                #edgecolor='black', #borda preta
                #linewidth=1.2, #espessura borda
                alpha=0.9 #transparência
                )
# Ajustar rótulos do eixo x
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Salvar o gráfico em alta resolução
plt.savefig('categorias_churn.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Evasão de clientes por tempo de contrtao em número absoluto

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Criar a coluna agrupada
df['Faixa_tempo_contrato'] = pd.cut(
    df['Tempo_contrato'],
    bins=range(0, df['Tempo_contrato'].max() + 3, 3),
    right=False,
    include_lowest=True
)

# Filtrar apenas churns: DataFrame apenas com valores de churn
df_churn = df[df['Churn'] == 1]

# Estilo
sns.set_theme(style='whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))
ax.set_title('Evasão de clientes por tempo de contrato (agrupado a cada 3 meses)', fontsize=18)
ax.set_ylabel('Total de clientes que deram churn', fontsize=14)
ax.set_xlabel('Tempo de contrato (meses)', fontsize=14)

# Plotagem
grafico = sns.countplot(
    data=df_churn,
    x='Faixa_tempo_contrato',
    palette='viridis',
    ax=ax
)

# Adicionar os valores no topo das barras
for barra in grafico.patches:
    altura = barra.get_height()
    if altura > 0:
        ax.text(
            barra.get_x() + barra.get_width() / 2,
            altura + 1,
            f'{int(altura)}',
            ha='center',
            fontsize=12,
            #fontweight='bold',
            color='black'
        )

# Ajustar rótulos do eixo x
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Salvar e exibir
plt.savefig('churn_por_tempo_contrato_agrupado.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#churn por custo do serviço oferecido
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Aplicando um tema
sns.set_theme(style='dark')

#Colunas a serem transformadas para linhas
servicos = [
    'Servico_Telefone', 'Multiplas_linhas',
       'Servico_internet', 'Seguranca_online', 'Backup_online',
       'Protecao_device', 'Suporte_internet', 'Streaming_tv',
       'Streaming_filme'
]

#Transposição
df_servicos = df.melt(
    id_vars=['Churn', 'Fatura_mensal'],
    value_vars=servicos,
    var_name='Servico',
    value_name='Ativo'
)
# Filtrar apenas os serviços ativos
df_ativos = df_servicos[df_servicos['Ativo']==1]

# Agora cria o gráfico: custo por serviço e status de churn
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(data=df_ativos, x='Servico', y='Fatura_mensal', hue='Churn',errorbar=None)

ax.set_title('Serviços por Custo Mensal e Churn', fontsize=18)
ax.set_ylabel('Custo mensal', fontsize=14)
ax.set_xlabel('Serviço')
plt.legend(title='Churn')
plt.tight_layout()




# Ajustar rótulos do eixo x
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.show()
```

## ✅ Conclusões e Insights
As principais características identificadas foram:
- 25,72% dos clientes na base cancelaram seus contratos;
- Os contratos com menos tempo apresentam um maior churn;
- Os serviços e tipos de contrato com maiores custos também apresentam maior churn;
- No geral, o custo médio do público que evadiu é de $74 e o que permaneceu é $61. Ou seja, a evasão está muito relacionada ao custo.
- Já para o tempo de contrato, foi possível identificar que o tempo médio ara o público que deu churn é de 17 meses e para os que permaneceram é de 37 meses. O maior público que cancelou está concentrado nos 3 primeiros meses.
- Eletronic check é o método de pagamento com o maior churn.
- Month-to-month é o tipo de contato com maior evasão.

## 💡 Recomendações

Algumas sugestões para tentar reduzir o percentual de churn:
- Negociar os tipos de contratos para que sejam mais longos, dando maior vantagem de custo para esse público.
- Entender o motivo do método "Eletronic check" estar relacionado a uma maior quantidade de churns.
- Revisão de preços a fim de nivelar com o mercado e nivelar os custos entre os próprios clientes da base que tem um custo mensal menor. E entender a motivação da diferença de valores.
